```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["apple","banana","orange","mango"])
y=np.array([30,40,50,60])
col=['red','blue','yellow','green']
plt.figure(figsize=(15, 8))
plt.barh(x,y,color=col)
plt.title('market')
plt.xlabel("fruite")
plt.ylabel("cost")
plt.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([1, 3, 4, 1, 3, 3, 2])
n, bins, patches = plt.hist(a, bins=4, edgecolor='black')
colors = ['pink', 'lightblue', 'lightgreen', 'lightcoral']
labels = ['DEV', 'OS', 'AA', 'D']
for patch, color, label in zip(patches, colors, labels):
    patch.set_facecolor(color)
    patch.set_label(label)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Values')
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([10, 30, 40,])
n,bins,patches=plt.hist(a,edgecolor='black',bins=3,range=(10,40))
colors=['red','yellow','blue']
labels=['apple','mango','grape']
for patch,color,label in zip(patches,colors,labels):
    patch.set_facecolor(color)
    patch.set_label(label)
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([20,30,40,40,20,20])
b=np.array([33,44,55,66,66,66])
c=np.array([88,33,22,11,22,11])
lable=['car','bike','cycle']
plt.hist((a,b,c),bins=4,label=lable,rwidth=0.4)
plt.legend()
plt.title('cost of vehical',fontsize=30)
plt.xlabel('cost',fontsize=30)
plt.ylabel('frequency',fontsize=30)
plt.show()
```
