```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
uniform_data=stats.uniform.rvs(size=10000,loc=0,scale=10)#rvs->random veriable 
#the values are bounded between 0 and 10
pd.DataFrame(uniform_data).plot(kind="density",figsize=(9,9),xlim=(-1,11))
#
```

```python
for x in range(-1,12,3):
    print("Density at x value"+str(x))
    print(stats.uniform.cdf(x,loc=0,scale=10))
 
```
