```python
import pandas as pd
data=pd.read_csv("C:/Users/dheve/OneDrive/Desktop/course/comedy(1).csv")
data.head(8)
```

```python
e={'UK':0,'USA':1,'N':2}
data[' Nationality']=data[' Nationality'].map(e)#map is used for read the content
d={'No':0,'Yes':1}
data['GO']=data['GO'].map(d)
data
```

```python
data.describe()
```

```python
features=['Age','Exp','Rank',' Nationality']
x=data[features]
y=data['GO']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
tree.plot_tree(dtree, feature_names=features)
#ginni refer to quality of spread
#value mean 60 % is yes and 70% is no
#ginni=1-(x/n)^2-(y/n)^2 for node its x=7 and y=6
```
