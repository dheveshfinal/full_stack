```python
#svm(support vactor machine)
#supervise machine learning alogoritham
#hyperplain
#Dmagin(sum)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv("C:/Users/dheve/OneDrive/Desktop/course/diabetes.csv")
```

```python
data
```

```python
non_zero = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
for coloumn in non_zero:
    data[coloumn] = data[coloumn].replace(0,np.NaN)
    mean = int(data[coloumn].mean(skipna = True))
    data[coloumn] = data[coloumn].replace(np.NaN,mean)
    print(data[coloumn])
```

```python
from sklearn.model_selection import train_test_split
X =data.iloc[:,0:8]
y =data.iloc[:,8]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0, stratify=y)
X.head()
#By using stratification, you ensure that the proportion of each class in the training set is similar to the proportion in the overall dataset.
#mean the train test prapotion is fixed they will not change
```

```python
#this give a z-score in floting points 
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
#test data also standardize in same way
#Standardize:z=(x-x(mean))/standard_deviation
#bring it to less veriation between each point
#Fit: It calculates the mean and standard deviation of each feature in the training dataset (X_train). This is important because these statistics will be used to standardize the data.
#Transform: It then applies the standardization process, transforming the training data such that each feature has a mean of 0 and a standard deviation of 1.
#Standardizing features by removing the mean and scaling to unit variance transforms the data so that it has a mean of 0 and a standard deviation of 1.
```

```python
X_train
```

```python
from sklearn import svm
svm1 = svm.SVC(kernel='linear', C = 0.01)
svm1.fit(X_train,y_train)
#SVM is a supervised machine learning algorithm used for classification tasks. It works by finding the hyperplane that best separates different classes in the feature space.
#A linear kernel means that the algorithm will find a linear hyperplane to separate the classes.
#The C parameter is a regularization parameter that controls the trade-off between achieving a low training error and a low testing error
```

```python
SVC(C=0.01, kernel='linear')
y_train_pred = svm1.predict(X_train)
y_test_pred = svm1.predict(X_test)
y_test_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_test_pred)
accuracy_score(y_test,y_test_pred)
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_test_pred})
df
```
