```python
#PCAPrincipal component Analysis
#we are using the PCA to reduce the resolusion of the image without changing the quality of the image
import pandas as pd
import numpy as np
```

```python
# Here we are using inbuilt dataset of scikit learn
from sklearn.datasets import load_breast_cancer
```

```python
# instantiating
cancer = load_breast_cancer(as_frame=True)#frame is used to remove the noise
#Frames refers to discrete individual Images or pictures when displayed sequentially in rapid succession
```

```python
df = cancer.frame
df
```

```python
# checking shape
print('Original Dataframe shape :',df.shape)
```

```python
# Input features
X = df[cancer['feature_names']]
print('Inputs Dataframe shape   :', X.shape)
```

```python
# Mean
X_mean = X.mean()
 
# Standard deviation
X_std = X.std()
 
# Standardization
Z = (X - X_mean) / X_std
```

```python
#covariance
c = Z.cov()#cov=summation((xi-x(mean)) (yi-y(mean)))/N-1
```

```python
c
```

```python
# Plot the covariance matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
sns.heatmap(c,annot=True)
plt.show()
```

```python
#linalg.eig is inbuilt funnction to calculate eigen values and eigen vectors
eigenvalues, eigenvectors = np.linalg.eig(c)#(linear_algebra)eigen vector is used for changing the horizontal and vertical
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
#Eigenvectors specify only the direction and do not change it under transformation.
# Each eigenvector has an associated eigenvalue, which represents the scaling factor in that direction.
#If the eigenvalue is greater than 1, the vector is scaled up, or "stretched." If it’s between 0 and 1, the vector is scaled down or "compressed."
```

```python
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
#cumulative sum is ratio of each priciple componets analysis sequentially
#variance tells us how much information or spread is captured in each direction (principal component) of the data.
```

```python
n_components = np.argmax(explained_var >= 0.50) + 1#argument matrices
n_components
#We often set a threshold like 50% (or higher) for explained variance to ensure we retain a meaningful portion of the data’s variability while reducing dimensionality
#argmax(argument function)-numpy(mandatory)
# Horizontal and verical component( orthogonal component)
# It returns the arguments for target function that returns maximum value from the target function
# represents the amount of variance captured by its corresponding eigenvector (or principal component). The total variance in the dataset is therefore the sum of all the eigenvalues.
```

```python
# PCA component or unit matrix
u = eigenvectors[:,:n_components]
pca_component = pd.DataFrame(u,
                             index = cancer['feature_names'],
                             columns = ['PC1','PC2']
                            )
 
# plotting heatmap
plt.figure(figsize =(5, 7))
sns.heatmap(pca_component)
plt.title('PCA Component')
plt.show()
```
