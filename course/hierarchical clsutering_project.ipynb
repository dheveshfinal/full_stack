```python
import pandas as pd
df=pd.read_csv(r"/content/Mall_Customers.csv")
df
# Hierarchical clustering is a machine learning algorithm that
# groups the data into tree of clusters
# It is an unsupervised machine learning algorithm
# Agglomerative clsutering-merge all clusters into one clusters
# Euclidean distance
# Label encoder method to convert categorical to numerical values
# USE CASE :WE ARE TRYING TO ANALYZE people s income and spending score TECHNIQUE by using hierarchical clustering method
```

```python
from sklearn import preprocessing
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
# unsupervised machine learning
```

```python
#step3:DATA VISUALIZATION
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(1 , figsize = (15 , 6))
n = 0
for x in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n += 1
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(df[x] , bins = 15)
    plt.title('Distplot of {}'.format(x))
plt.show()
# Sub(1,2,2),first row,Two plots,second plot
#(1,3,n=1)=first row,three plots,first plot
```

```python
#Step4: Label encoder method:categorical values to numerical values
label_encoder = preprocessing.LabelEncoder()

df['Gender'] = label_encoder.fit_transform(df['Gender'])
df.head()
```

```python
#step5: Dendogram is a tree like diagram that represents a hierarchical structure or the results of a clustering algorithm
# ward in python:Euclidean distance
plt.figure(1, figsize = (16 ,8))
dendrogram = sch.dendrogram(sch.linkage(df, method  = "ward"))#ward how the cluster is merged

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
#we infer that we get good hierarchical clustering for better data visualization(single clsuter formed)
```
